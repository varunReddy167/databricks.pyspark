```
df=spark.range(2)
df.show()
```

```
from pyspark.sql.functions import current_date, date_format
df=df.withColumn('current_date',current_date())
df.show()
df.printSchema()
```

```
from pyspark.sql.functions import  date_format, to_date
df1=df.withColumn('current_date', date_format(df.current_date, 'yyyy.dd.MM'))

df1.show()
df1.printSchema()
```

```
from pyspark.sql.functions import  date_format, to_date
df2=df1.withColumn('current_date',to_date(df1.current_date, 'yyyy.dd.MM'))
df2.show()
df2.printSchema()
```

```
#DATEDIFF
from pyspark.sql.functions import datediff

df=spark.createDataFrame([('2023-07-3','2023-04-3')],['startdate','enddate'])
df.show()
```

```
from pyspark.sql.functions import datediff,months_between,add_months,date_add,year,month
df=df.withColumn('datefiffc',datediff(df.startdate,df.enddate))
df=df.withColumn('betwen_months',months_between(df.startdate,df.enddate))
df=df.withColumn('addmonths',add_months(df.startdate,3))  
df=df.withColumn('submonths',add_months(df.startdate,-3)) 
df=df.withColumn('date_add',date_add(df.startdate,10)) 
df=df.withColumn('year',year(df.startdate)) 
df=df.withColumn('month',month(df.startdate)) 


df.show()
```
