```
data=[('varun', 'reddy'), ('Vinay', 'reddy')]
```

```
#rdd cretation
rdd=spark.sparkContext.parallelize(data)
```

```
#rdd is immutable so it assiged to another rdd
rdd1=rdd.map(lambda x:x + (x[0]+ x[1],))
rdd1.collect()
```

```
rdd1.count()

rdds=rdd.map(lambda x:str(x).split(','))
rdds.collect()
```

```
#dataframe to rdd
data=[('varun', 'reddy'), ('Vinay', 'reddy')]
df=spark.createDataFrame(data,['fn','ln'])
```

Flat_map()

```
data=[('varun reddy'), ('Vinay reddy')]
```

```
rdd=spark.sparkContext.parallelize(data)
```

```
rdd.count()
```

```

#in map transformation when you give count of 2 inputs the output will be also count of 2
rdd1=rdd.map(lambda x:x.split(' '))
rdd1.collect()

for item in rdd1.collect():
    print(item)
```

```
rdd1.count()
```

```
#in falt_map transformation it faltten the data and provies the 2 count to 4 count outpit when you split
rdd2=rdd.flatMap(lambda x:x.split(' '))
rdd2.collect()

for item in rdd2.collect():
    print(item)
```

```
rdd2.count()
```
